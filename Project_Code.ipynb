```python
#K=1, T=1
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("Traffic Light")
m.Params.LogToConsole = 1

R = m.addVars(2, name="R")
G = m.addVars(2, name="G")


m.setObjective(G[0] + G[1], GRB.MINIMIZE)

m.addConstr(R[0] + R[1] + G[0] + G[1] == 120)
m.addConstr(R[0] >= 5)
m.addConstr(R[1] >= 5)
m.addConstr(G[0] >= 5)
m.addConstr(G[1] >= 5)
m.addConstr(R[0] + R[1] >= G[0] + G[1])
m.addConstr(R[0] == R[1])
m.addConstr(G[0] == G[1])

m.optimize()

print("Optimal value = ", m.objVal)
for v in m.getVars():
    print('%s = %g' % (v.varName, v.x))
```
